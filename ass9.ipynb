```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mylabels = ["Studying", "Sleeping", "Playing", "Hobby", "Friends&Family"]
plt.pie(np.array([33,30,5,18,14]), labels=mylabels)
plt.show()
```

```python
data = {'a' : [75, 82, 88, 95, 67, 84, 90, 72, 79, 91, 68, 78, 86, 98, 80],
        'b' : [60, 70, 62, 83, 95, 71, 68, 87, 75, 56, 72, 80, 65, 69, 74]
        }
df = pd.DataFrame(data=data)
df.boxplot()

```

```python
grades_data = {
    "Students" : ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    "Grades" : [85, 70, 90, 60, 75]
}

df = pd.DataFrame(grades_data)
plt.bar(df['Students'], df['Grades'])
```

```python
scores = [75, 80, 85, 90, 60, 70, 65, 95, 88, 92, 78, 82, 85, 70, 75, 92, 68, 72, 88, 76]
num_bins = 10
plt.hist(scores, bins=num_bins, edgecolor='k')
```
