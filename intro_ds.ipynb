```python
import pandas as pd

df = pd.read_csv('dataset1.csv')
df
```

```python
import matplotlib.pyplot as plt

country_stats = {}
all_countries = set(df['country']) 
for country in all_countries:
    country_stats[country] = {
        'data' : df[df["country"] == country]
    }

num_subplots = len(country_stats.items())
num_rows = 30
num_columns = num_subplots // num_rows + 1

fig, axes = plt.subplots(num_rows, num_columns, figsize=(15, 100))
p = 0
axes = axes.ravel()

for key, value in country_stats.items():
    ax = axes[p]
    ax.plot(value['data']['year'], value['data']['lifeExp'])
    ax.set_title(key)
    ax.set_xlabel('Years')
    ax.set_ylabel('Life Exp')
    p += 1

plt.tight_layout()
plt.show()
```

```python
num_subplots = len(country_stats.items())
num_rows = 30
num_columns = num_subplots // num_rows + 1

fig, axes = plt.subplots(num_rows, num_columns, figsize=(15, 100))
p = 0
axes = axes.ravel()

for key, value in country_stats.items():
    ax = axes[p]
    ax.plot(value['data']['year'], value['data']['gdpPercap'])
    ax.set_title(key)
    ax.set_xlabel('Years')
    ax.set_ylabel('Life Exp')
    p += 1

plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9, wspace=0.4, hspace=0.4)

plt.show()
```

```python
p = 0
for key, value in country_stats.items():
    if(p < 5):
        plt.bar(value['data']['year'], value['data']['gdpPercap'], label=key)
    else:
        break
    p += 1

plt.legend()
plt.show()
```
